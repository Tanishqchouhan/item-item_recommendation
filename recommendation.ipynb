```python
# Combine items into lists by order number
grouped = allT.groupby('CustomerID')['StockCode'].apply(list)
print(type(grouped))
grouped.head()
```

```python
# Combine items into lists by order number
grouped = allT.groupby('CustomerID')['StockCode'].apply(list)
print(type(grouped))
grouped.head()
```

```python
# One-hot encode data in pandas dataframe
te = TransactionEncoder()
te_ary = te.fit(grouped).transform(grouped)
te__ary  = te_ary.astype("int")
transaction_group = pd.DataFrame(te__ary, columns=te.columns_)
transaction_group
```

```python
print(transaction_group.shape)
final  = transaction_group[0:100]
final
print(final.shape)
```

```python
hero =  final.corr()
type(hero)
```

```python
c = hero.abs()
s = c.unstack()
so = s.sort_values(kind="quicksort")

print (so)
```
